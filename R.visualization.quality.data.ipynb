```r
ratings<-c(2,4,3,3,2,1,1,2,3,4,2,3,3,4,1,3,2,1,4,3,2,4)
```

```r
lengths(ratings)
```

```r
summary(ratings)
```

```r
hist(ratings,col='blue')
```

```r
hist(ratings,col='blue',prob=TRUE)
```

```r
hist(ratings,col='blue',prob=FALSE)
```

```r
hist(ratings,col='grey')
lines(density(ratings),col='red')
```

```r
hist(ratings,col='grey',probability=TRUE)
lines(density(ratings),col='red')
```

```r
hist(ratings,col='grey',probability=FALSE)
lines(density(ratings),col='red')
```

```r
hist(ratings,col='grey',probability=TRUE,ylim=c(0,0.6))
lines(density(ratings),col='red')
```

```r
boxplot(ratings)
```

```r
plot(ratings)
```

```r
stripchart(ratings)
```

```r
stripchart(ratings,col='red')
```

```r
stripchart(ratings,col='red',pch=15)
```

```r
stripchart(ratings,col='red',pch=15,method='jitter')
```

```r
stripchart(ratings,col='red',pch=15,method='stack')
```
